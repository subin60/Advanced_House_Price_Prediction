```python
train_dataset = pd.read_csv('train_dataset1.csv')
test_dataset = pd.read_csv('test_dataset1.csv')
```

```python
train_dataset
```

```python
X = train_dataset.drop(['SalePrice','Id'], axis=1)
```

```python

y = train_dataset['SalePrice']
```

```python
X
```

```python
test_dataset = test_dataset.drop(['Id'], axis=1)
```

```python
train_dataset.shape, test_dataset.shape
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(X, y)
```

```python
selected_feat = X.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((X.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
X = X[selected_feat]
test_dataset = test_dataset[selected_feat]
```

```python
X
```

```python
test_dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0)
```

```python
pipeline = Pipeline(steps=[
    ('rf', RandomForestRegressor())
])

param_grid = {
    'rf__n_estimators': [10, 50, 100],
    'rf__max_depth': [None, 10, 30],
    'rf__min_samples_split': [2, 5, 10],
    'rf__min_samples_leaf': [1, 2, 4]
}

search = GridSearchCV(pipeline, param_grid, cv=10, verbose=1, n_jobs=-1)
search.fit(X_train, y_train)

y_pred = search.predict(X_test)

# Calculate evaluation metrics for regression
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

# Calculate RMSE between logarithms of predicted and observed values
#log_rmse = np.sqrt(mean_squared_error(np.log(y), np.log(y_pred)))
log_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
print("Logarithmic RMSE:", log_rmse)
```

```python
# Apply the inverse transformation (exponential) on the predicted and actual target variables
y_val_pred = np.exp(y_pred)
y_val = np.exp(y_test)

# Calculate the Root Mean Squared Error (RMSE) between the inverse-transformed predicted and actual target variables
log_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Logarithmic RMSE:", log_rmse)
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

final_pipe = make_pipeline(
    SelectKBest(f_regression),
    RandomForestRegressor()
)

pipe_params = {
    'selectkbest__k': [1, 10, 20],
    'randomforestregressor__max_depth': [10, 20, 40],
    'randomforestregressor__min_samples_split': [2, 3, 5],
    'randomforestregressor__min_samples_leaf': [1, 2, 5]
}

search = GridSearchCV(final_pipe, pipe_params, cv = 10, scoring='neg_root_mean_squared_error', 
                   error_score='raise')

search.fit(X_train, y_train)

y_pred = search.predict(X_test)

# Calculate evaluation metrics for regression
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

# Calculate RMSE between logarithms of predicted and observed values
#log_rmse = np.sqrt(mean_squared_error(np.log(y), np.log(y_pred)))
log_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
print("Logarithmic RMSE:", log_rmse)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

final_pipe = make_pipeline(
    MLPRegressor(random_state=0)
)

pipe_params = {
    'mlpregressor__hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'mlpregressor__activation': ['relu', 'tanh'],
    'mlpregressor__solver': ['adam', 'lbfgs'],
    'mlpregressor__alpha': [0.0001, 0.001, 0.01],
    'mlpregressor__learning_rate': ['constant', 'adaptive'],
    'mlpregressor__max_iter': [5000]
}

search = GridSearchCV(final_pipe, pipe_params, cv=10, scoring='neg_root_mean_squared_error', 
                      error_score='raise')

search.fit(X_train, y_train)

y_pred = search.predict(X_test)

# Calculate evaluation metrics for regression
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

log_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
best_params = search.best_params_

print("Best Parameters:", best_params)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
print("Logarithmic RMSE:", log_rmse)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Define a pipeline with StandardScaler and MLPRegressor
pipeline = make_pipeline(StandardScaler(), MLPRegressor(random_state=0))

# Define the hyperparameters to search
param_grid = {
    'mlpregressor__hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'mlpregressor__activation': ['logistic', 'tanh', 'relu', 'identity'],
    'mlpregressor__solver': ['lbfgs', 'sgd', 'adam'],
    'mlpregressor__learning_rate': ['constant', 'invscaling', 'adaptive'],
    'mlpregressor__max_iter': [5000],
}

# Set up GridSearchCV with the pipeline and hyperparameters
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_root_mean_squared_error', n_jobs=-1, verbose=2)

# Fit the GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters found by GridSearchCV:")
print(grid_search.best_params_)

# Evaluate the model on the test data
y_pred = grid_search.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```
