```python
def replace_cat_feature(train_dataset, test_dataset, features_nan_train, features_nan_test):
    train_data = train_dataset.copy()
    test_data = test_dataset.copy()

    # Replace missing categorical features with 'Missing' in train_data
    train_data[features_nan_train] = train_data[features_nan_train].fillna('Missing')
    
    # Replace missing categorical features with 'Missing' in test_data
    test_data[features_nan_test] = test_data[features_nan_test].fillna('Missing')

    return train_data, test_data

# Find categorical features with missing values in both train and test datasets
features_nan_train = [feature for feature in train_dataset.columns if train_dataset[feature].isnull().sum()>1 and train_dataset[feature].dtypes=='O']
features_nan_test = [feature for feature in test_dataset.columns if test_dataset[feature].isnull().sum()>1 and test_dataset[feature].dtypes=='O']

# Replace missing values in categorical features in both train and test datasets
train_dataset, test_dataset = replace_cat_feature(train_dataset, test_dataset, features_nan_train, features_nan_test)

# Check if missing values are replaced in train_dataset
print(train_dataset[features_nan_train].isnull().sum())

# Check if missing values are replaced in test_dataset
print(test_dataset[features_nan_test].isnull().sum())
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan_train = [feature for feature in train_dataset.columns if train_dataset[feature].isnull().sum()>1 and train_dataset[feature].dtypes !='O']
numerical_with_nan_test = [feature for feature in test_dataset.columns if test_dataset[feature].isnull().sum()>1 and test_dataset[feature].dtypes !='O']

for feature in numerical_with_nan_train:
    print("{}: {}% missing values".format(feature,np.round(train_dataset[feature].isnull().mean(),4)))

for feature in numerical_with_nan_test:    
    print("{}: {}% missing values".format(feature,np.round(test_dataset[feature].isnull().mean(),4)))
```

```python
def replace_num_missing_values(train_dataset, test_dataset, numerical_with_nan_train, numerical_with_nan_test):
    train_data = train_dataset.copy()
    test_data = test_dataset.copy()
    
    # Replace missing numerical values in train_data
    for feature in numerical_with_nan_train:
        median_value = train_data[feature].median()

        # Create a new feature to capture NaN values
        train_data[feature + 'nan'] = np.where(train_data[feature].isnull(), 1, 0)
        train_data[feature].fillna(median_value, inplace=True)

    # Replace missing numerical values in test_data
    for feature in numerical_with_nan_test:
        median_value = test_data[feature].median()

        # Create a new feature to capture NaN values
        test_data[feature + 'nan'] = np.where(test_data[feature].isnull(), 1, 0)
        test_data[feature].fillna(median_value, inplace=True)

    return train_data, test_data

# Find numerical features with missing values in both train and test datasets
numerical_with_nan_train = [feature for feature in train_dataset.columns if train_dataset[feature].isnull().sum()>1 and train_dataset[feature].dtypes!='O']
numerical_with_nan_test = [feature for feature in test_dataset.columns if test_dataset[feature].isnull().sum()>1 and test_dataset[feature].dtypes!='O']

# Replace missing values in numerical features in both train and test datasets
train_dataset, test_dataset = replace_num_missing_values(train_dataset, test_dataset, numerical_with_nan_train, numerical_with_nan_test)

# Check for missing values in the train and test datasets
print("Train dataset missing values:")
print(train_dataset[numerical_with_nan_train].isnull().sum())

print("\nTest dataset missing values:")
print(test_dataset[numerical_with_nan_test].isnull().sum())
```

```python
def convert_years_to_age(train_dataset, year_features, sold_year_feature):
    data = train_dataset.copy()
    
    for feature in year_features:
        data[feature] = data[sold_year_feature] - data[feature]
    
    return data

# List of features to be converted
year_features = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']
sold_year_feature = 'YrSold'


train_dataset = convert_years_to_age(train_dataset, year_features, sold_year_feature)
test_dataset = convert_years_to_age(test_dataset, year_features, sold_year_feature)

# Check the transformed features in the train and test datasets
print("Train dataset transformed year features:")
print(train_dataset[year_features].head())

print("\nTest dataset transformed year features:")
print(test_dataset[year_features].head())
```

```python
def apply_log_transform(train_dataset, num_features):
    data = train_dataset.copy()
    
    for feature in num_features:
        data[feature] = np.log(data[feature])
    
    return data

# List of numerical features to be transformed
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']


train_dataset = apply_log_transform(train_dataset, num_features)


num_features.remove('SalePrice')
test_dataset = apply_log_transform(test_dataset, num_features)

# Check the transformed features in the train and test datasets
print("Train dataset transformed numerical features:")
print(train_dataset[num_features].head())

print("\nTest dataset transformed numerical features:")
print(test_dataset[num_features].head())
```

```python


def handle_rare_vars(dataset, categorical_features):
    for feature in categorical_features:
        temp = dataset.groupby(feature)['Id'].count() / len(dataset)
        temp_df = temp[temp > 0.01].index
        dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'Rare_var')
    
    return dataset


train_categorical_features = [feature for feature in train_dataset.columns if train_dataset[feature].dtype == 'O']
test_categorical_features = [feature for feature in test_dataset.columns if test_dataset[feature].dtype == 'O']

train_dataset = handle_rare_vars(train_dataset, train_categorical_features)
test_dataset = handle_rare_vars(test_dataset, test_categorical_features)

# Check the transformed features in the train and test datasets
print("Train dataset transformed categorical features:")
print(train_dataset[train_categorical_features].head())

print("\nTest dataset transformed categorical features:")
print(test_dataset[test_categorical_features].head())
```

```python
categorical_features = [feature for feature in train_dataset.columns if train_dataset[feature].dtype == 'O']

mappings = {}
for feature in categorical_features:
    labels_ordered = train_dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered = {k: i for i, k in enumerate(labels_ordered, 0)}
    train_dataset[feature] = train_dataset[feature].map(labels_ordered)
    mappings[feature] = labels_ordered

for feature in categorical_features:
    # Map the test dataset categories and fill missing ones with the most frequent category in the train dataset
    most_frequent_category = train_dataset[feature].mode().iloc[0]
    test_dataset[feature] = test_dataset[feature].map(lambda x: mappings[feature].get(x, most_frequent_category))
```

```python
print("Train dataset transformed categorical features:")
print(train_dataset[categorical_features].head())

print("\nTest dataset transformed categorical features:")
print(test_dataset[categorical_features].head())
```

```python
feature_scale = [feature for feature in train_dataset.columns if feature not in ['Id', 'SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the train dataset
train_dataset[feature_scale] = scaler.fit_transform(train_dataset[feature_scale])

# Transform the test dataset using the scaler fitted on the train dataset
test_dataset[feature_scale] = scaler.transform(test_dataset[feature_scale])

# Check the transformed features in the train and test datasets
print("Train dataset after feature scaling:")
print(train_dataset[feature_scale].head())

print("\nTest dataset after feature scaling:")
print(test_dataset[feature_scale].head())
```
